Prove by mathematical induction sum of i * factorial(i) = factorial(i+1) - 1

$\sum_{i=1}^{n} i! \cdot i = (n+1)! - 1$

base case: $n=1$; LHS = $\sum_{i=1}^{1} i! \cdot i = 1! \cdot 1 = 1$ RHS = $(1+1)! - 1 = 2! - 1 = 1$ LHS = RHS, so base case holds. 

Let's assume this relation holds for $(n-1)$; $\sum_{i=1}^{n-1} i! \cdot i = n! - 1$ - assumption 

So, prove for $n$: LHS = $\sum_{i=1}^{n} i! \cdot i = \sum_{i=1}^{n-1} i! \cdot i + n! \cdot n$ $= n! - 1 + n! \cdot n$ (from (1)) 

$= n!(1+n) - 1$

$= (n+1)n! - 1 = (n+1)! - 1 = \text{RHS}$

Hence proved by induction.

Problem #1:
Write a function to perform integer division without using either the / or * operators. Find a fast way to do it.

Solution:
Words algorithm:
We know that if b has to be divided into a size parts, we ll have, b = a*m + c, where we ll have m parts and c left. You can keep adding a to itself until it is less than b, maintain an variable i which increments in every iteration. 



```python
print("Problem #1 implementation")

b = int(input("Enter the divisor"))
a = int(input("Enter the dividend"))
temp = a
m = 1 #Quotient
while(b > temp):
    temp += a
    m += 1
if(temp  > b):
    m -= 1
print(f"Dividing {b} by {a} you get {m}")
```

**Problem #2** : There are 25 horses. At most, 5 horses can race together at a time. You must determine the fastest, second fastest, and third fastest horses. Find the minimum number of races in which this can be done.

**Solution**: Pick horses randomly and then race em in sets of 5, note their completion times. Once you have all the horses race times, sort em descending and choose first 3. This will only require 5 races!

**Sub-problem**: What if you don't have a stopwatch.

**Solution**: Get the top 5 horses by racing em into groups of 5. This will require 5 races. Now have a 6th race to get the fastest horse. We need one more 7th race in which we race the second and third placed horses in final race with the second and third placed horses in the group of winning horse. Whoever finishes top two in this race is second and third fastest overall.

3n+1 Problem (https://cheenta.com/the-3n1-problem/)
 Its a conjecture where you start arbitary from any number n, if you 
 - encounter an even number, you do n/2
 - encounter odd number you do 3n+1
 You ll always reach 1, after arbitary number of iterations.

```python
n = int(input("Enter the natural number from where you want to start"))
print(f"Starting from {n}")
while(n > 1):
    if(n % 2 == 0):
        n = n/2
    else:
        n = 3*n + 1
    print(f"{n}, ")
```

**The Trip Problem (Splitwise algorithm)**:
You and many friends go to a trip, multiple people pay different bills and at end of the trip you need to find the total and settle all balances.

**Input**: You have number of people on trip 'n' and an array of money spent by each person.
**Solution**:
- Start by calculating total expense by adding each expense. 
- Then find the share of each person by dividing total expense by number of people.
- Next subtract the share from each element in array.
- Sort the array.
- Now, we start from left, take the leftmost element, sum it with right most, if it comes out to be zero or positive, the left element is the amount owed by left person to right. If non-zero sum, save it in temp variable. Move the left pointer by one, repeat sum
